```python
import speech_recognition as sr   # recognizer - it it speech or not, listener - analyze the speech and convert to audio if it is speech
```

```python
r = sr.Recognizer()
r
```

```python
audio = sr.AudioFile('/content/harvard.wav')
with audio as source:
  audio1 = r.record(source)
```

```python
with audio as source:
  audio2 = r.listen(source)
```

```python
# without using listener
# Now that the audio is recorded, you can use a recognizer to convert it to text.

text = r.recognize_google(audio1) # Using audio1 which contains the recorded audio data
print(f"Recognized text: {text}")
```
